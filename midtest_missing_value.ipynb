```python
### memilih baris yang tidak kosong
ttc_update = ttc_update[~ttc_update['EmbarkedCity'].isnull()].reset_index(drop = True)
```

```python
ttc_update.info()
```

Kolom Embarked sudah dihilangkan missing valuesnya. Sehingga tersisa 889 rows yang sudah tidak ada missing valuesnya lagi. Jadi, kolom yang sisa adalah 'Age'.

```python
ttc_update[ttc_update['Age'].isnull()]
```

Berarti, terdapat sekitar 20% baris yang mengalami missing value di kolom Age.


cek median umur dari passenger di Titanic, berdasarkan tingkat keselamatan, jenis kelamin, dan kota keberangkatan

```python
ttc_update.groupby(['Survived', 'Sex', 'EmbarkedCity'])['Age'].median()
```

```python
ttc_update.groupby(['Survived', 'Sex', 'EmbarkedCity'])['Age'].median().reset_index()
```

```python
median_umur = ttc_update.groupby(['Survived', 'Sex', 'EmbarkedCity'])['Age'].median().reset_index()
median_umur
```

```python
### Ganti Nama Kolom 

median_umur['MedianAge'] = median_umur['Age']
```

```python
median_umur
```

```python
median_umur.drop(['Age'], axis = 'columns', inplace = True)
median_umur
```

Mengisi missing value dengan median umur yang sudh dicari

```python
ttc_update[ttc_update['Age'].isnull()].head(3)
```

```python
median_umur
```

Dapat dilakukan dengan MERGE 
Merge TTC UPDATE dengan MEDIAN_UMUR

```python
ttc_update.merge(median_umur, how = 'left')
```

```python
ttc_update = ttc_update.merge(median_umur, how = 'left')
```

```python
ttc_update.head(6)
```

Terdapat kolom baru di ttc_update. Cek kembali ttc_update yang mengalami missing value pada kolom AGE


```python
ttc_update[ttc_update['Age'].isnull()].head(3)
```

Setiap nilai nan di kolom AGE dapat diganti dengan MedianAge untuk tahap selanjutnya

```python
ttc_update['Age'] = ttc_update['Age'].fillna(ttc_update['MedianAge'])
```
