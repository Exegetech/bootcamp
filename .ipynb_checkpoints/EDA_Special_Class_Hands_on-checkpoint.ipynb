## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# bank marketing campaign data
# https://www.kaggle.com/datasets/prakharrathi25/banking-dataset-marketing-targets/versions/1?resource=download&select=train.csv
bank = pd.read_csv('bank_marketing.csv')
bank.head()
```

Bank marketing campaign data:
1. ID : ID of client
2. age (numeric)
3. job : type of job (categorical: "admin.","blue-collar","entrepreneur","housemaid","management","retired","self-employed","services","student","technician","unemployed","unknown")
4. marital : marital status (categorical: "divorced","married","single"; note: "divorced" means divorced or widowed)
5. education (categorical: "primary","secondary","tertiary,"unknown")
6. default: has credit in default? (categorical: "no","yes")
7. balance : has money in account? (numeric)
8. housing: has housing loan? (categorical: "no","yes")
9. loan: has personal loan? (categorical: "no","yes")
10. contact: contact communication type (categorical: "cellular","telephone")
11. month: last contact month of year (categorical: "jan", "feb", "mar", â€¦, "nov", "dec")
12. day: last contact day of the week (categorical: "mon","tue","wed","thu","fri")
13. duration: last contact duration, in seconds (numeric). Important note: this attribute highly affects the output target (e.g., if duration=0 then y="no"). Yet, the duration is not known before a call is performed. Also, after the end of the call y is obviously known. Thus, this input should only be included for benchmark purposes and should be discarded if the intention is to have a realistic predictive model.
14. campaign: number of contacts performed during this campaign and for this client (numeric, includes last contact)
15. pdays: number of days that passed by after the client was last contacted from a previous campaign (numeric; -1 means client was not previously contacted)
16. previous: number of contacts performed before this campaign and for this client (numeric)
17. poutcome: outcome of the previous marketing campaign (categorical: "failure","unknown","other","success")
18. y - has the client subscribed a term deposit? (binary: "yes","no")

```python
# numerical variable stats summary
bank[numericals].describe()
```

Oservation:

## Univariate Analysis

```python
# adjust the figure size for better readability
plt.figure(figsize=(15,5))

# plotting
features = numericals
for i in range(0, len(features)):
    plt.subplot(1, len(features), i+1)
    sns.boxplot(y=bank[features[i]], color='red')
    plt.tight_layout()
```

Observation:
1. Several columns have outliers: age, balance, duration, campaign, pdyas and previous
2. For modelling objective, we need to explore if log transformation is useful for those columns, esepcially for balance, duration, campaign, pday and previous.

```python
# adjust the figure size for better readability
plt.figure(figsize=(15,6))

features = numericals
for i in range(0, len(features)):
    plt.subplot(2, len(features)//2 + 1, i+1)
    sns.kdeplot(x=bank[features[i]], color='skyblue')
    plt.xlabel(features[i])
    plt.tight_layout()
```

Observation:
1. ID column is PK, just ignore it
2. NO column distributed nicely as either normal or symmetric distribution
3. Confirmed as in outlier observation: balance, duration, campaign, pdays and previous are positively skewed. Will test log transforming them.
4. Day has multimodal distribution

```python
bank[numericals].describe()
```

```python
enforce_positive_cols = ['balance','duration','campaign','pdays','previous']
bank_positive_cols = bank.copy()
for col in enforce_positive_cols:
  if col == 'pdays':
    bank_positive_cols[col] = bank_positive_cols[col] + 2 # min value is -1
  else :
     bank_positive_cols[col] = bank_positive_cols[col] + 1 # min value is 0
```

```python
# adjust the figure size for better readability
plt.figure(figsize=(15,6))

features = enforce_positive_cols
for i in range(0, len(features)):
    plt.subplot(2, len(features)//2 + 1, i+1)
    sns.kdeplot(x=np.log(bank_positive_cols[features[i]]), color='skyblue')
    plt.xlabel(features[i])
    plt.tight_layout()
```

1. balance and duration are better after log transformed! (hence we can perform log transform for modelling later--not in this EDA)
2. while campaign, pdays and previous are not the case (hence we can leave as is for modelling later)

```python
# adjust the figure size for better readability
plt.figure(figsize=(15,9))

features = categoricals
for i in np.arange(1, len(features)+1):
    plt.subplot(3, len(features)//3+1, i)
    sns.countplot(x=bank[features[i-1]], color='green')
    plt.xlabel(features[i-1])
    plt.tight_layout()
```

# Traffic Dataset

```python
# load traffic data
traffic = pd.read_csv('/content/drive/MyDrive/Dibimbing/Unified Materials/EDA II Extra Class/Metro_Interstate_Traffic_Volume.csv')
traffic.head()
```

```python
# general dataframe info
traffic.info()
```

Observation
1. No missing values detected
2. All columns, except date_time, have the correct dtypes already
3. date_time needs to be converted to datetime format 

```python
# duplicated check
traffic.duplicated().sum()
```

```python
# drop duplicates
traffic = traffic.drop_duplicates()
```

```python
# convert date_time to datetime format
traffic['date_time'] = pd.to_datetime(traffic['date_time'])
traffic.info()
```

Deep-dive questions:
1. month-basis timeseries of traffic volume
2. Weekly seasonality pattern Weekly seasonality pattern (Average of daily traffic)
3. Weekdays vs Weekend traffic profile (boxplot)/
4. Top weather description in causing traffic volume
5. Daily Seasonality pattern (per jam, dibreakdown weekend vs not)

```python
# 1. month-basis timeseries of traffic volume
# Hint: Truncate to month level, then sum aggregate the traffic
# Note: need to convert the month column as string
```

```python
# 2. Weekly seasonality pattern (Average of daily traffic)

# create date column
traffic['date'] = traffic['date_time'].dt.to_period('D')

# sum aggregate traffic per date
traffic_day = traffic.groupby('date').agg(
    traffic = ('traffic_volume','sum')
).reset_index()

traffic_day.head()
```

```python
#  create day_name column
traffic_day['date'] = traffic_day['date'].astype(str)
traffic_day['date'] = pd.to_datetime(traffic_day['date'])
traffic_day['day_name'] = traffic_day['date'].dt.day_name()

# aggregation (average) group by day_name
traffic_daily = traffic_day.groupby('day_name').agg(
    avg_traffic = ('traffic', 'mean')
).reset_index()

# day index for proper ordering in visualization 
day_index = {
    'Monday':1,
    'Tuesday':2,
    'Wednesday':3,
    'Thursday':4,
    'Friday':5, 
    'Saturday':6,
    'Sunday':7
}

traffic_daily['day_index'] = traffic_daily['day_name'].map(day_index)
traffic_daily = traffic_daily.sort_values('day_index', ignore_index=True) 
traffic_daily
```

```python
# visualize it
plt.figure(figsize=(8,5))
sns.barplot(data=traffic_daily, x='day_name', y='avg_traffic')
plt.xticks(rotation=90)
plt.show()
```

* Weekdays have higher traffic (relatively constant)
* Sunday has the least traffic

```python
# 3. weekdays vs weekend traffic profile

# leverage traffic_day to create is_weekend flag
traffic_day['is_weekend'] = traffic_day['day_name'].apply(lambda x: 1 if x in ['Saturday','Sunday'] else 0)
```

```python
# craete boxplot
plt.figure(figsize=(6,5))
sns.boxplot(data=traffic_day, x='is_weekend', y='traffic')
```

Further validation of the previous finding: traffic in weekend is lower

```python
# 4. Top weather description in causing traffic volume
# Try yourself!
```

```python
# 5. Daily Seasonality pattern (per jam, dibreakdown weekend vs not)

# create is_weekend flag on traffic dataframe
traffic['day_name'] = traffic['date_time'].dt.day_name()
traffic['is_weekend'] = traffic['day_name'].apply(lambda x: 1 if x in ['Saturday', 'Sunday'] else 0)

# truncate date_time up to hour level
traffic['hour'] = traffic['date_time'].dt.to_period('H')

# aggregate the hourly traffic
traffic_hourly = traffic.groupby(['is_weekend','hour']).agg(
    traffic = ('traffic_volume','sum')
).reset_index()

traffic_hourly.head()
```

```python
# extract hour digit
traffic_hourly['hour'] = traffic_hourly['hour'].astype(str)
traffic_hourly['hour'] = pd.to_datetime(traffic_hourly['hour'])
traffic_hourly['hour_digit'] = traffic_hourly['hour'].dt.hour
traffic_hourly.head()
```

```python
# barplot visualization
traffic_hourly_sim = traffic_hourly.groupby(['is_weekend','hour_digit']).agg(
    avg_traffic = ('traffic','mean')
).reset_index()

plt.figure(figsize=(20,6))
sns.barplot(data=traffic_hourly_sim, x='hour_digit', y='avg_traffic', hue='is_weekend')
plt.xticks(rotation=90)
plt.title("Average Hourly Traffic Volume: Weekdays (blue) vs Weekend (orange)", fontsize=14)
plt.show()
```

Very insightful!
