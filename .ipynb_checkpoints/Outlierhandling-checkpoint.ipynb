```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.DataFrame([[1,2,3,4,5,6],[25,30,55,40,35,260],[5000,6500,4500,7000,7500,14000]]).transpose()
df.columns = ['ID','UmurCustomer','TotalHarga']
df
```

> Contoh ada sebuah DataFrame yang merupakan tabel penjualan.

```python
fig, axarr = plt.subplots(1,3, figsize=(10, 4))
for x in df.columns:
    axarr[df.columns.get_loc(x)].boxplot(df[x])
    axarr[df.columns.get_loc(x)].set_xlabel(x)
plt.suptitle("Outliers checking on numeric columns")
fig.tight_layout(pad=1)
plt.show() 
```

>Saat melakukan outliers checking menggunakan boxplots didapat 2 kolom yang terdapat outliers yaitu UmurCustomer dan TotalHarga.<br>
1.Untuk outlier pada UmurCustomer, nilai tersebut merupakan nilai yang tidak valid dikarenakan tidak ada manusia yang berumur sampai 260 tahun dan alangkah baiknya jika nilai tersebut didrop.<br>
2.Untuk outlier pada TotalHarga, nilai tersebut tidak pasti merupakan nilai yang tidak valid dikarenakan mungkin bisa saja ada customer yang berbelanja sangat banyak. Jika nilai tersebut merupakan nilai valid, sebaiknya tidak didrop saat melakukan proses EDA tetapi boleh di drop atau tidak sebelum modelling tergantung analisa masing-masing.
