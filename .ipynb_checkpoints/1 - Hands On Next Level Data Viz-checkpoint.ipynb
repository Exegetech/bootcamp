```python
# Ignore warnings

import warnings

warnings.simplefilter(action = "ignore", category = FutureWarning)
# Import the dependencies

# import the necessary Python libraries Numpy, Pandas and Matplotlib with their usual shorthand notation.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
import random
```

```python
path = 'mpg_update.csv'
mpg = pd.read_csv(path, index_col=0)
mpg.head(5)
```

##### Menampilkan presentase record
Untuk menampilkan presentase record dapat menggunakan pie chart, tetapi sebelum itu perlu melakukan preprocessing untuk mendapatkan presentase nya.

Cara mudah untuk mendapatkan presentase dapat menggunakan function value_counts(normalize=True)*100

```python
percentage=mpg['mpg_level'].value_counts(normalize=True)*100
percentage
```

Highlight bagian medium dengan mengatur warna, apply medium dengan warna orange dan darkgrey untuk bagian lainnya.

```python
color = ["orange", "darkgrey", "darkgrey"]
plt.pie(percentage.values, labels=percentage.index, startangle=90, autopct='%1.0f%%', explode=(0.1,0,0), colors = color, wedgeprops = { 'linewidth' : 1 , 'edgecolor' : 'white'} )

plt.title("the percentage of records")
plt.xlabel("mpg level")
plt.show()
```

### Fill Between

Fill between digunakan untuk mengisi warna suatu area

Sebagai contoh, terdapat garis lengkung di bawah ini

```python
# make data
x = np.linspace(0, np.pi * 4, 300)
y = np.sin(x)

# plot sin and line
plt.plot(x, y, color='red')
plt.axhline(0) 
```


Warnai bagian tertentu dengan menggunakan function fill_between.
- Atur bagian hline dan y yang nilainya > 0 dengan warna hijau
- gunakan parameter where dan color, plt.fill_between(x, y, where=(), color='')

> *use where to create boolean masks determining where to fill*

Selanjutnya coba atur bagian hline dan y, 
- yang nilainya > 0 dengan warna hijau, 
- dan < 0 dengan warna merah

```python
# make data
x = np.linspace(0, np.pi * 4, 300)
y = np.sin(x)

# plot sin and line
plt.plot(x, y, color='red')
plt.axhline(0)

plt.fill_between(x, y, where=(y > 0), color='green')
plt.fill_between(x, y, where=(y < 0), color='red', hatch='/', alpha=.3)
```

### Discrete dan Continous Coloring

Digunakan untuk meng-encode numerical continous variable

```python
sns.scatterplot(x="horsepower", y="mpg", 
            data=mpg)
```

Untuk membuat discrete coloring di scatter plot, dapat menambahkan hue

#### Coba bandigkan perbedaannya

gunakan plt.subplots dan set axis

> fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

sns.scatterplot(x="horsepower", y="mpg", 
            data=mpg, 
            hue="cylinders",
            ax=ax1)

cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)

sns.scatterplot(x="horsepower", y="acceleration", 
            data=mpg, 
            hue="mpg",
            cmap=cmap)
```

Visualisasikan continous data dengan bantuan color bar

```python
plt.scatter(mpg['horsepower'], mpg['acceleration'], c=mpg['mpg'], s=30, cmap=cmap)
```

### Facet Grid

FacetGrid digunakan untuk memvisualisasikan distribusi satu variabel serta hubungan antara beberapa variabel secara terpisah
> Simplenya, Facet grid seperti Hue, tapi dipisah pisah menjadi column

Untuk membuat facetgrid, gunakan sns.FacetGrid dan map_dataframe. Langkah:
- sns.FacetGrid(data, col="column")
- fig.map_dataframe(jenis plot, x_value, y_value)

Contoh di bawah kita ingin memvisualisasaikan scatter plot menggunakan variable horsepower dan acceleration berdasarkan origin, tapi dipisah origin satu per satu

Di dalam facet grid, dapat juga menambahkan hue, fig = sns.FacetGrid(mpg, col="origin", hue="mpg_level")

```python
fig = sns.FacetGrid(mpg, col="origin", hue="mpg_level")
fig.map_dataframe(sns.scatterplot, x="horsepower", y="acceleration")

fig.add_legend()
fig.set_xlabels('Horse Power')
fig.set_ylabels('Acceleration')
```

### Line
Line digunakan untuk memotong sumbu x atau y dan biasanya digunakan untuk menambahkan threshold dalam visualisasi

> Untuk menambahkan line horizontal dapat menggunakan plt.axhline(). 

```python
fig = sns.boxplot(x="origin", y="mpg", data=mpg)
```

Sekarang coba kita lihat penggunaan bahan bakar berdasarkan origin apakah lebih dari rata rata mobil yang diproduksi atau tidak?
- lakukan perhitungan mean
- buat boxplot
- buat garis threshold mean menggunakan plt.axhline()

Atur linestyle menjadi dashed, dengan menambahkan parameter **linestyle='dashed'**

```python
plt.subplots(figsize=(8, 6))

average_mpg = mpg['mpg'].mean()

fig = sns.boxplot(x="origin", y="mpg", data=mpg)
plt.axhline(average_mpg, color='r', linestyle='dashed', linewidth=2)
```

### Annotate

Annotate biasanya digunakan untuk menambahkan keterangan berupa text atau 

```python
fig = sns.boxplot(x="origin", y="mpg", data=mpg)
```

contoh, kita ingin masukkan text "spontan uhuy" di box plot
- tambahkan plt.annotate("text")
- atur posisi plt.annotate("text", xy=(x, y))

Sekarang tambahkan informasi median di masing masing box
- generate nilai median dengan df.groupby().median()
- input annotate untuk semua box menggunakan perulangan for dan tambahkan annotate

```python
median = mpg.groupby(['origin'],sort=False)['mpg'].median()
median
```

lakukan perulangan for dan tambahkan annotate untuk menambahkan informasi median untuk semua box

```python
fig = sns.boxplot(x="origin", y="mpg", data=mpg)
for i in range(len(median)):
  fig.annotate(str(median[i]), xy=(i, median[i]) , horizontalalignment='center')

# fig.annotate("18,5", xy=(0, 18.5) , horizontalalignment='center')
# fig.annotate("31,6", xy=(1, 31.6) , horizontalalignment='center')
# fig.annotate(str(median[i]), xy=(i, median[i]) , horizontalalignment='center')
```

```python
order_product = pd.DataFrame({
  'product': ['product A', 'product B', 'product C', 'product D', 'product E'],
  'rata-rata repeat order':  [3, 4, 6, 5, 4],
  'jumlah konsumen': [55, 70, 45, 60, 30]
})
order_product
```

```python
sns.scatterplot(x='jumlah konsumen', y='rata-rata repeat order', data=order_product)
```

Tambahkan annotate menggunakan plt.annotate()

```python
value_repeat_order = order_product['rata-rata repeat order'].values
value_jumlah_konsumen = order_product['jumlah konsumen'].values
```

```python
sns.scatterplot(x='jumlah konsumen', y='rata-rata repeat order', data=order_product)
plt.annotate('product A', xy=(55, 3))
plt.annotate('product B', xy=(70, 4))
plt.annotate('product B', xy=(45, 6))
```

Tambahkan annotate menggunakan perulangan, supaya lebih presisi valuenya dan tidak menginisialisasi 1 per 1

```python
sns.scatterplot(x='jumlah konsumen', y='rata-rata repeat order', data=order_product)
value_repeat_order = order_product['rata-rata repeat order'].values
value_jumlah_konsumen = order_product['jumlah konsumen'].values

for i, txt in enumerate(order_product['product']):
    plt.annotate(txt, xy=(value_jumlah_konsumen[i], value_repeat_order[i]))
```

```python
sns.scatterplot(x='jumlah konsumen', y='rata-rata repeat order', data=order_product)
value_repeat_order = order_product['rata-rata repeat order'].values
value_jumlah_konsumen = order_product['jumlah konsumen'].values

for i, txt in enumerate(order_product['product']):
    plt.annotate(txt, xy=(value_jumlah_konsumen[i], value_repeat_order[i]))

max_repeat_order = max(order_product['rata-rata repeat order'].values)
jumlah_konsumen_of_max_repeat_order = order_product['jumlah konsumen'].loc[order_product['rata-rata repeat order'] == max_repeat_order].values[0]

x_annotate = jumlah_konsumen_of_max_repeat_order - 0.1
y_annotate =  max_repeat_order - 0.1

plt.annotate('repeat order rate paling tinggi', xy=(x_annotate, y_annotate), 
            xytext=(0.30, 0.6), textcoords='axes fraction',
            arrowprops=dict(facecolor='black'),
            horizontalalignment='center'
            )
```
