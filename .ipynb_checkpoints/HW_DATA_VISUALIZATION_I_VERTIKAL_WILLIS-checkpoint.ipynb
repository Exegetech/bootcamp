# PART 1: TELCO CUSTOMER CHURN

> ## Data Introduction
This dataset contain informations about customers that churn and not churned in a telco company. Below is the column informations:<Br>
1.customerID = customer unique ID.<Br>
2.gender = customer gender (M/F).<br>
3.SeniorCitizen = old / young customer.<br>
4.Partner = either a customer has partners or not.<br>
5.Dependents = either a customer has dependents or not.<br>
6.tenure = how long the customer subscribed.<br>
7.MultipleLines = either a customer using multiple lines or not.<br>
8.InternetService = either a customer using InternetService lines or not.<br>
9.OnlineSecurity = either a customer has OnlineSecurity or not.<br>
10.OnlineBackup = either a customer has OnlineBackup or not.<br>
11.DeviceProtection = either a customer has DeviceProtection or not.<br>
12.TechSupport = either a customer has TechSupport or not.<br>
13.StreamingTV = either a customer has StreamingTV or not.<br>
14.StreamingMovies = either a customer has StreamingMovie or not.<br>
15.Contract = types of contract.<Br>
16.PaperlessBilling = either a customer has PaperlessBilling or not.<br>
17.PaymentMethod = types of the payment method.<Br>
18.MonthlyCharges = how much charges per month.<br>
19.TotalCharges = total charges of all time.<br>
20.Churn = either a customer churn or not.<br>

```python
#feature engineering & feature selection for simpler EDA
df['IsSingle'] = df[['Partner','Dependents']].apply(lambda x: 'Yes' if list(x).count('Yes') > 0 else 'No', axis=1)
df['Security'] = df[['OnlineSecurity','DeviceProtection']].apply(lambda x: 'Yes' if list(x).count('Yes') > 0 else 'No', axis=1)
df['Supports'] = df[['OnlineBackup','TechSupport']].apply(lambda x: 'Yes' if list(x).count('Yes') > 0 else 'No', axis=1)
df['Entertain'] = df[['StreamingTV','StreamingMovies']].apply(lambda x: 'Yes' if list(x).count('Yes') > 0 else 'No', axis=1)
df['Services'] = df[['PhoneService','InternetService']].apply(lambda x: 'Both' if list(x).count('No') == 0 else 'Internet Only' if x[0] == 'No' else 'Phone Only', axis=1)
df.drop(columns=['customerID','Partner','Dependents','OnlineSecurity','DeviceProtection','OnlineBackup','TechSupport','StreamingTV','StreamingMovies','PhoneService'], inplace=True)
```

3. Supports = true if customer has either online backup or tech support.<br>
4. Entertain = true if customer has either streaming tv or streaming movies.<br>
5. Services = check if customer has both internet service and phone service.<br>

## Exploratory data analysis with data visualization.
***

```python
#create a function to plot probability.
def prob_plot(df,colom,x):
    means = df[colom].mean()
    medians = df[colom].median()
    data = df[df.Churn == 'Yes'][colom].astype('float64')
    data1 = df[df.Churn == 'No'][colom].astype('float64')
    
    kde = gaussian_kde(data)
    kde1 = gaussian_kde(data1)
    dist_space = np.linspace( min(data), max(data), 200)
    dist_space1 = np.linspace( min(data1), max(data1), 200)
    axarr[x].plot( dist_space, kde(dist_space), label='Churned', color='orange' )
    axarr[x].plot( dist_space1, kde1(dist_space1), label='Not churn', color='blue')
    axarr[x].axvline(x = means, linestyle = '--', color='g', label='Mean')
    axarr[x].axvline(x = medians, linestyle = '--', color='r', label='Median')
    axarr[x].set_title('Probability', fontweight='bold', size=12)
    axarr[x].set(ylabel = 'Probability', xlabel = colom)
    axarr[x].legend()
```

### MonthlyCharges analysis

```python
fig, axarr = plt.subplots(1,2, figsize=(12, 4))
g = sns.histplot(data=df, x = 'MonthlyCharges',hue='Churn', ax=axarr[0], multiple='stack')
axarr[0].set_title('Distribution', fontweight='bold', size=12)
prob_plot(df,'MonthlyCharges',1)
axarr[1].legend(loc='upper right')
plt.show()
```

>It seems that the more expensive the monthly charges the more likely customers will churn... but this information that these plots give is little bit blurry, we need to do more analysis.

```python
#Let's conduct a quick hyphothesis testing with Monthlycharges between the group that either has security, supports and entertain and the group that has nothing.
g1 = df[(df.Security == 'Yes') | (df.Supports == 'Yes') | (df.Entertain == 'Yes')]['MonthlyCharges']
g2 = df[(df.Security == 'No') & (df.Supports == 'No') & (df.Entertain == 'No')]['MonthlyCharges']
#g1 = group that either has benefit on security, supports or entertain.
#g2 = group that has no benefit at all.
stat.ttest_ind(g1,g2, equal_var=False)
```

```python
a = df.groupby('Security')[['MonthlyCharges']].agg(Security = ('MonthlyCharges','mean')).transpose()
b = df.groupby('Supports')[['MonthlyCharges']].agg(Supports = ('MonthlyCharges','mean')).transpose()
c = df.groupby('Entertain')[['MonthlyCharges']].agg(Entertain = ('MonthlyCharges','mean')).transpose()
pd.concat([a,b,c]).plot(kind='bar', stacked=True, rot = 0, width=0.4)
plt.title('MonthlyCharges average between benefits', weight='bold')
plt.legend(loc='upper left')
plt.show()
```

>After conducting t-test and plotting the benefit features vs monthly charges, it is true that the more benefits the customer use the higher the monthly charges is.

## Data cleaning & data wrangling
***

```python
df = pd.DataFrame()
for y in ['17','18','19']:
    link = 'https://raw.githubusercontent.com/vertikalwil/Data-Analyst/main/datasets/'
    p = 1
    for x in ['jan','feb','mar','apr','mei','jun','jul','agus','sept','okto','nov','des']:
        csvv = pd.read_csv(link + x + y +'.csv', encoding='unicode_escape')
        csvv['tahun'] = '20'+ y
        csvv['bulan'] = p
        p += 1
        df = pd.concat([df,csvv])
```

```python
df = df.iloc[:,0:7].copy().convert_dtypes().reset_index(drop=True)
df
```

```python
df.info()
```

```python
df['jumlah_penumpang'] = df['jumlah_penumpang'].astype('string').str.replace(',','').str.replace('.','',regex=False).astype('float64')
df['tahun'] = df['tahun'].astype('float64')
df['bulan'] = df['bulan'].astype('float64')
df['penumpang/bus'] = (df.jumlah_penumpang / df.jumlah_bus).astype('float64')
```

```python
df[df.type_operasi.isnull() | df.jumlah_bus.isnull()]
```

```python
#drop NaN values.
df.drop(index=[263,279,280,368], inplace=True)
```

```python
df.info()
```

```python
df['daerah_operasi'] = df['daerah_operasi'].str.replace(' ','')
```

```python
fig, axarr = plt.subplots(1,2, figsize=(10, 4))
y = 0
for x in df[['jumlah_bus','jumlah_penumpang']].columns:
    axarr[y].boxplot(df[x])
    axarr[y].set_xlabel(x)
    y += 1
    if y == 3:
        break
plt.suptitle("Outliers checking on numeric columns")
fig.tight_layout(pad=1)
plt.show() 
```

> banyaknya outliers pada numerical kolom, maka semua analisa average akan menggunakan nilai median.

```python
df['rata2_jumlah_bus'] = df.groupby(['tahun','bulan'])['jumlah_bus'].transform('median').round().astype('float64')
df['rata2_jumlah_penumpang'] = df.groupby(['tahun','bulan'])['jumlah_penumpang'].transform('median').round().astype('float64')
df['rata2_penumpang/bus'] = df.groupby(['tahun','bulan'])['penumpang/bus'].transform('median').round().astype('float64')
```

```python
df.info()
```

```python
df2017 = df[df.tahun == 2017]
df2018 = df[df.tahun == 2018]
df2019 = df[df.tahun == 2019]
```

## Exploratory data analysis
***

```python
fig, axarr = plt.subplots(1,2, figsize=(10, 4))
data = df[['tahun','bulan','rata2_jumlah_bus','rata2_jumlah_penumpang']].drop_duplicates()
a = sns.lineplot(data=data, x='bulan', y='rata2_jumlah_bus', hue='tahun', ax=axarr[0], legend=False, palette="PuBuGn_d")
b = sns.lineplot(data=data, x='bulan', y='rata2_jumlah_penumpang', hue='tahun', ax=axarr[1], legend=True, palette="PuBuGn_d")
b.get_legend().remove()
handles, labels = b.get_legend_handles_labels()
fig.legend(handles, [int(float(s)) for s in labels], loc='upper right', ncol=3, bbox_to_anchor=(.68, 0.1))
axarr[0].set_title('Rata-rata jumlah bus', weight='bold')
axarr[1].set_title('Rata-rata jumlah penumpang', weight='bold')
fig.tight_layout(pad=1)
```

>Terlihat dari 2 grafik diatas bahwasannya jumlah penumpang maupun jumlah bus menurun pada sekitar pertengahan tahun yaitu pada bulan 6. Jika dilihat perbandingan jumlah bus antara tahun 2017 - 2019 nilainya cukup seimbang namun jumlah penumpangnya membludak sangat jauh pada tahun 2017. Kemungkinan terjadi overcrowded yang berlebihan pada tahun 2017 yang mengakibatkan penumpang menurun drastis di tahun berikutnya yaitu 2018 dan 2019.

```python
fig, axarr = plt.subplots(1,3, figsize=(15, 6))
x = df2017.groupby('daerah_operasi')['jumlah_penumpang'].agg('mean').sort_values(ascending=False).head(10)
y = df2017.groupby('daerah_operasi')['jumlah_bus'].agg('mean').sort_values(ascending=False).head(10)
z = df2017.groupby('daerah_operasi')['penumpang/bus'].agg('mean').sort_values(ascending=False).head(10)
x.plot(kind='bar',rot = 45, ax=axarr[0], title='Rata-rata penumpang', color='green')
y.plot(kind='bar',rot = 45, ax=axarr[1], title='Rata-rata bus', color='brown')
z.plot(kind='bar',rot = 45, ax=axarr[2], title='Rata-rata penumpang/bus')
plt.suptitle("Perbandingan daerah operasi 2017", weight='bold')
plt.show()
```

```python
#kalkulasi nilai std deviation.
z.std()
```

>Rata-rata penumpang tertinggi ada pada rute 7, namun untuk jumlah bus tertinggi pada rute 5. Untuk rata-rata penumpang/bus bisa dilihat bahwa terjadi overcrowded pada rute 7, ini cocok dengan grafik rata-rata penumpang dimana rute 7 merupakan yang tertinggi namun untuk penempatan armada bus nya malah menempati peringkat ke 7. Ini menandakan bahwa terjadi salah penempatan bus atau mis-management. Untuk rata-rata penumpang/bus, nilai variance atau perbedaan antar rute masih terlihat cukup jomplang. Semakin jomplang nilai grafik tersebut maka semakin jelek management penempatan bus. Terakhir, terlalu banyak penumpang (ribuan) yang menempati 1 bus. Supply dan demand sangat berbeda jauh pada tahun ini.
