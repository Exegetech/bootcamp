Kolom-kolom dari dataset 'diamonds.csv' memiliki definisi sebagai berikut:
- price: price in US dollars ($326--$18,823)
- carat: weight of the diamond (0.2--5.01)
- cut: quality of the cut (Fair, Good, Very Good, Premium, Ideal)
- color: diamond colour, from J (worst) to D (best)
- clarity: a measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))
- x: length in mm 
- y: width in mm 
- z: depth in mm 
- depth: total depth percentage = z / mean(x, y) = 2 * z / (x + y) 
- table: width of top of diamond relative to widest point

#### Nomor 1:

```python
df = df.convert_dtypes() #auto convert data type.
df.info()
print('\nBisa dilihat bahwa kolom yang mengandung nilai NA / kosong adalah carat dengan jumlah 125 baris per 53,940 baris atau sekitar 0.2%' )
```

```python
#Menghitung nilai median berdasarkan nilai gabungan atau group dari cut, color dan clarity.
dfmedian = df.groupby(['cut','color','clarity'])[['carat']].agg(NilaiMedianCarat = ('carat','median')).reset_index() #mencari nilai median per 'cut','color' dan 'clarity'
dfmedian
```

```python
df = df.merge(dfmedian, how='left') #left join / merge nilai median dengan data df awal.
df
```

```python
df[df['carat'].isnull()].head()
```

```python
df['carat'] = df['carat'].fillna(df['NilaiMedianCarat']) #mengisi nilai kosong dengan nilai median.
df.drop(columns=['NilaiMedianCarat'], inplace=True) #membuang kolom nilai median karena sudah tidak diperlukan.
df.info()
```

Done, tidak ada lagi missing values. Saya memilih metode mengisi nilai NA menggunakan nilai median (group by cut, color, clarity) pada kolom carat karena nilai NA nya sangat sedikit yaitu 0.2% sehingga penginputan nilai median dapat dikatakan nilai yang cukup akurat dan terpercaya.

#### Nomor 2:

Penjelasan clarity yang sudah disebutkan diawal : "clarity: a measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))" Karena kategori clarity bersifat ordinal (rank / berurut) maka kita akan melakukan rank encoding.

```python
ranks = {
    'I1' : 1,
    'SI2' : 2,
    'SI1' : 3,
    'VS2' : 4,
    'VS1' : 5,
    'VVS2': 6,
    'VVS1': 7,
    'IF'  : 8
}
```

```python
df['clarity'] = df.clarity.map(ranks)
df.head(10)
```

#### Nomor 3:

```python
fig, axarr = plt.subplots(1,3 , figsize=(15,5))
axarr[0].boxplot(df.x, labels='X')
axarr[1].boxplot(df.y, labels='Y')
axarr[2].boxplot(df.z, labels='Z')
plt.show()
```

Melihat data boxplot diatas, maka bisa ditarik kesimpulan: <Br>
1.Perlu dilakukan pengecekan dan analisa untuk nilai 0 pada kolom x,y dan z karena ukuran dimensi tidak mungkin 0.<br>
2.Pada kolom x,y dan z terdapat sekumpulan data outlier yang nilainya tidak terlalu jauh dari batas atas / batas bawah. Kemungkinan data tersebut merupakan data real, namun perlu dilakukan analisa lebih lanjut.<Br>
3.Terdapat 3 nilai outlier yang sangat extreme (nilai berada sangat jauh dari batas atas) pada kolom y dan z. Kemungkinan data tersebut salah (typo, kesalahan sistem, dll), namun tetap perlu dilakukan analisa lanjutan dari mana data tersebut muncul.

```python
def outlier_detection(col):
    dff = pd.DataFrame(col)
    IQR = np.quantile(col, 0.75) - np.quantile(col, 0.25)
    lower_bound = np.quantile(col, 0.25) - (1.5 * IQR)
    upper_bound = np.quantile(col, 0.75) + (1.5 * IQR)
    dff['status'] = dff.apply(lambda x: 'Outlier-ExtremeTopValue' if x[0] > upper_bound else 'Outlier-ExtremeBottomValue' if x[0] < lower_bound else 'Not Outliers', axis=1)
    return dff    
    
```

```python
#Menampilkan 10 nilai outlier pada kolom x
dfoutlierx = outlier_detection(df.x)
dfoutlierx = dfoutlierx[dfoutlierx.status != 'Not Outliers']
dfoutlierx.head(10)
```

```python
#Menampilkan 10 nilai outlier pada kolom y
dfoutliery = outlier_detection(df.y)
dfoutliery = dfoutliery[dfoutliery.status != 'Not Outliers']
dfoutliery.head(10)
```

```python
#Menampilkan 10 nilai outlier pada kolom z
dfoutlierz = outlier_detection(df.z)
dfoutlierz = dfoutlierz[dfoutlierz.status != 'Not Outliers']
dfoutlierz.head(10)
```

```python
print(f'Ada sebanyak {len(dfoutlierx)} jumlah baris outlier pada kolom x')
print(f'Ada sebanyak {len(dfoutliery)} jumlah baris outlier pada kolom y')
print(f'Ada sebanyak {len(dfoutlierz)} jumlah baris outlier pada kolom z')
```
