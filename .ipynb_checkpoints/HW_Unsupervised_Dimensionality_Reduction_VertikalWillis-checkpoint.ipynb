## Import Package

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_squared_error as mse
```

## Load Dataset

```python
from sklearn.datasets import load_boston
boston_dataset = load_boston()
boston = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
boston['MEDV'] = boston_dataset.target
boston.head()
```

```python
df = boston
```

## Data Preprocessing tanpa reduksi dimensi (20 point)

### Log Transform Skew Features

```python
plt.figure(figsize=(15,12))

features = df.columns
for i in np.arange(1, len(features)+1):
    plt.subplot(2, len(features)//2, i)
    sns.histplot(df[features[i-1]],color='skyblue', kde=True, edgecolor='none')
    plt.xticks(rotation=10)
    plt.xlabel(features[i-1])
plt.suptitle('Distribution Plot', size=19, weight='bold')
plt.tight_layout(pad = 1)
```

```python
#enforce positive cols karena log transform tidak bisa menerima nilai dibawah 1
enforce_positive_cols = ['CRIM','ZN','INDUS','NOX','B']
for col in enforce_positive_cols:
    df[col] = df[col] + 1  
```

```python
#log transform skew features
skew = ['CRIM','ZN','INDUS','NOX','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']
for x in skew:
    df[x] = np.log(df[x])
```

### Scaling

```python
df = pd.DataFrame(MinMaxScaler().fit_transform(df), columns=df.columns)
```

```python
x = df.drop(['MEDV'], axis = 1)
y = df[['MEDV']]
```

## Data Pre-Processing dengan tambahan PCA Prosess (40)

```python
df1 = df
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=7)
pca.fit(x)
pcs = pca.transform(x)

pdf = pd.DataFrame(data = pcs, columns = ['pc1', 'pc2', 'pc3', 'pc4','pc5','pc6','pc7'])
pdf['MEDV'] = y
pdf.head()
```

```python
print('Explained variance:', pca.explained_variance_)
print('Explained variance ratio:', pca.explained_variance_ratio_)
```

```python
cumsum = np.cumsum(pca.explained_variance_ratio_*100)
cumsum
```

>Disini kita memerlukan PC dengan minimal 80% yaitu ada pada minimum PC4 (82%). Tetapi jika kita menggunakan sampai PC5, cumsumnya bertambah 5% menjadi 87% dan jika menggunakan PC6 kenaikan cumsumnya hanya bertambah 2% dari PC5. Walau bisa menggunakan PC 4, saya lebih memilih PC5 karena kenaikan cumsumnya yang cukup berasa kenaikannya dari PC4 dan menggunakan 5 features dari 13 features yang ada menurut saya sudah sangat bagus.

```python
x_pca = pdf.drop(['pc6','pc7','MEDV'], axis=1)
y_pca = pdf[['MEDV']]
```

## Regression Model (20 point)

```python
X_train, X_test,Y_train,Y_test = train_test_split(x, y,test_size = 0.25, random_state = 42)
X_train_pca, X_test_pca,Y_train_pca,Y_test_pca = train_test_split(x_pca,y_pca,test_size = 0.25, random_state = 42)
```

```python
X_train = X_train.to_numpy()
Y_train = Y_train.to_numpy()
Y_train = Y_train.reshape(len(Y_train),)

X_test = X_test.to_numpy()
Y_test = Y_test.to_numpy()
Y_test = Y_test.reshape(len(Y_test),)

X_train_pca = X_train_pca.to_numpy()
Y_train_pca = Y_train_pca.to_numpy()
Y_train_pca = Y_train_pca.reshape(len(Y_train_pca),)

X_test_pca = X_test_pca.to_numpy()
Y_test_pca = Y_test_pca.to_numpy()
Y_test_pca = Y_test_pca.reshape(len(Y_test_pca),)
```

```python
#saya akan menggunakan simple linear regression saja.
reg = LinearRegression().fit(X_train, Y_train)
regPCA = LinearRegression().fit(X_train_pca, Y_train_pca)
```

```python
y_predict = reg.predict(X_test)
y_predictPCA = regPCA.predict(X_test_pca)
```

```python
maePCA = mae(Y_test_pca, y_predictPCA)
maePCA
```

```python
mae = mae(Y_test ,y_predict)
mae
```

```python
rmsePCA = np.sqrt(mse(Y_test_pca ,y_predictPCA))
rmsePCA
```

```python
rmse = np.sqrt(mse(Y_test ,y_predict))
rmse
```

## Evaluasi (20)

Bandingkan hasil menggunakan PCA dan tidak menggunakan PCA

```python
result = pd.DataFrame([mae,maePCA,rmsePCA,rmse]).transpose()
result.columns = ['mae','maePCA','rmsePCA','rmse']
```

```python
#hasil PCA adalah menggunakan PCA dengan 5 components.
result
```

> Didapat dari hasil diatas bahwa dengan menggunakan PCA akurasinya lebih rendah dari tanpa menggunakan PCA (sekitar +- 10%). Namun perlu diingat bahwa PCA hanya menggunakan 5 kolom predictor dimana tanpa menggunakan PCA kolom predictornya ada 13. Jumlah kolom yang digunakan bisa dipangkas lebih dari 50% namun hasil akurasinya hanya berbeda sedikit.

>Note untuk perbandingan component PCA saya hanya membandingkan sampai 7 componen karena pada component ke-7 saja sudah didapat informasi diatas 90%. Component ideal yang saya rasa terbaik digunakan adalah 5 component (information loss hanya sekitar 13%) karena jika menggunakan 6 component akurasinya hanya naik sedikit dan jika menggunakan 4 component akurasinya turun lumayan banyak.
