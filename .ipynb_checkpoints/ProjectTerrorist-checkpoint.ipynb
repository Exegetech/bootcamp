```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("globalterror.csv", delimiter=",", encoding='latin1')
df
```

```python
df.info(verbose=True, show_counts=True)
```

```python
for x in df.columns:
    if df[x].count() < len(df.index) / 2:
        df.drop(columns=[x], inplace=True)
df.drop(columns=['guncertain1','summary','eventid','country','region','provstate','specificity','vicinity','crit1','crit2','crit3','doubtterr','multiple','attacktype1','targtype1','targsubtype1','natlty1','individual','nperpcap','claimed','weaptype1','weapsubtype1','nkillter','nkillus','nwoundus','nwoundte','property','ishostkid','scite1','INT_LOG','INT_IDEO','INT_MISC','INT_ANY'], inplace=True)
df.drop(73127,axis=0,inplace=True)
df.drop_duplicates(inplace=True)      
```

```python
for x in ['city','targsubtype1_txt','corp1','target1','natlty1_txt','weapsubtype1_txt','weapdetail']:
    df[x].fillna('Unknown', inplace=True)
for x in ['latitude','longitude','nkill','nwound']:
    df[x].fillna(0, inplace=True)
df.nperps.fillna(1, inplace=True)
df.nperps.replace(-99,1, inplace=True)
```

```python
df = df.convert_dtypes()
df.info()
```

```python
def level_check(x):
    if (x.nkill == 0 and x.nwound == 0):
        return'Low'
    elif (x.nkill <= 10 and x.nwound <= 50 ):
        return 'Moderate'
    elif (x.nkill <= 50 and x.nwound <= 100):
        return 'High'
    else:
        return 'Extreme'

def group_check(x):
    if (x.nperps == 1):
        return 'Individual'
    elif (x.nperps <= 10):
        return 'Small'
    elif (x.nperps <= 50):
        return 'Medium'
    elif (x.nperps <= 150):
        return 'Big'
    else:
        return 'Massive'
```

```python
df['TerrorLevel'] = df.apply(lambda x: level_check(x), axis=1)
df.TerrorLevel.value_counts()
```

```python
df['PerpGroup'] = df.apply(lambda x: group_check(x), axis=1)
df['causalities'] = df['nwound'] + df['nkill']
```
