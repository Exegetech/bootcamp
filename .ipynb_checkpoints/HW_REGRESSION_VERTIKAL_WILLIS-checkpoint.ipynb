# Data Pre-Processing

```python
plt.figure(figsize=(15,12))

features = df.columns
for i in np.arange(1, len(features)+1):
    plt.subplot(2, len(features)//2, i)
    sns.histplot(df[features[i-1]],color='skyblue', kde=True, edgecolor='none')
    plt.xticks(rotation=10)
    plt.xlabel(features[i-1])
plt.suptitle('Distribution Plot', size=19, weight='bold')
plt.tight_layout(pad = 1)
```

```python
from sklearn.model_selection import train_test_split

x = df.drop(columns='medv')
y = df[['medv']]

# First splitting: pretrain and test
x_pretrain, x_test, y_pretrain, y_test = train_test_split(x, y, test_size=0.20, random_state=42)

# Second splitting: train and validation
x_train, x_validation, y_train, y_validation = train_test_split(x_pretrain, y_pretrain, test_size=0.20, random_state=42)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif 
from statsmodels.tools.tools import add_constant

X = add_constant(x_train)

vif_df = pd.DataFrame([vif(X.values, i) 
               for i in range(X.shape[1])], 
              index=X.columns).reset_index()
vif_df.columns = ['feature','vif_score']
vif_df = vif_df.loc[vif_df.feature!='const']
vif_df.sort_values('vif_score', ascending=False)
```

```python
total_train = pd.concat([x_train, y_train], axis=1)
plt.figure(figsize=(15,9))
sns.heatmap(total_train.corr(), annot=True, )
plt.show()
```

>After playing with the features I choose to drop feature 'rad','dis' and 'nox' because these feature have very high multicollinearity and keep 'indus' and 'tax' features. I also consider to drop 'age' but dropping this feature doesn't change much the vif value so I think this feature is not really redundant and can provide some good informations to the model. Here's the detail:<br>
1.'tax' and 'rad' have high multicollinearity and I choose to drop 'rad' because 'tax' has higher correlation to 'medv'.<br>
2.'indus','dis','nox' and 'age' also have high multicollinearity and I choose to keep 'indus' because it has highest correlation to target. So I drop 'dis' and 'nox' but not 'age' because of the reason I explained above.<Br>
3. I tried all the combinations and find out that the best features to drop are 'rad','dis' and 'nox'.

```python
x_train = x_train.drop(columns=['rad','dis','nox'])
x_validation = x_validation.drop(columns=['rad','dis','nox'])
x_test = x_test.drop(columns=['rad','dis','nox'])
```

```python
X = add_constant(x_train)

vif_df = pd.DataFrame([vif(X.values, i) 
               for i in range(X.shape[1])], 
              index=X.columns).reset_index()
vif_df.columns = ['feature','vif_score']
vif_df = vif_df.loc[vif_df.feature!='const']
vif_df.sort_values('vif_score', ascending=False)
```

```python
total_train = pd.concat([x_train, y_train], axis=1)
plt.figure(figsize=(15,9))
sns.heatmap(total_train.corr(), annot=True, )
plt.show()
```

>All seems good!

# Modelling

## Ridge Regression

```python
from sklearn.linear_model import Ridge

# train the model
x_train = x_train.to_numpy()
y_train = y_train.to_numpy()
y_train = y_train.reshape(len(y_train),)

# define the model
# 4 models with 4 different alphas (lambda)
ridge_reg_pointzeroone = Ridge(alpha=0.01, random_state=42)
ridge_reg_pointone = Ridge(alpha=0.1, random_state=42)
ridge_reg_one = Ridge(alpha=1, random_state=42)
ridge_reg_ten = Ridge(alpha=10, random_state=42)

# fit the model (training)
ridge_reg_pointzeroone.fit(x_train, y_train)
ridge_reg_pointone.fit(x_train, y_train)
ridge_reg_one.fit(x_train, y_train)
ridge_reg_ten.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error as mape
from sklearn.metrics import mean_absolute_error as mae

x_validation = x_validation.to_numpy()
y_validation = y_validation.to_numpy()
y_validation = y_validation.reshape(len(y_validation),)

alphas = [0.01, 0.1, 1., 10]
models = [ridge_reg_pointzeroone,
          ridge_reg_pointone,
          ridge_reg_one,
          ridge_reg_ten]

for model, alpha in zip(models, alphas):
    y_predict_validation = model.predict(x_validation)
    rmse = np.sqrt(mean_squared_error(y_validation ,y_predict_validation))
    mape1 = mape(y_validation ,y_predict_validation)
    mae1 = mae(y_validation ,y_predict_validation)
    print(f'RMSE of Ridge regression model with alpha = {alpha} is {rmse}')
    print(f'MAPE of Ridge regression model with alpha = {alpha} is {mape1}')
    print(f'MAE of Ridge regression model with alpha = {alpha} is {mae1}')
    print('--------------------------------------------------')
```

> Ridge Regression with alpha 10 is the best model with the lowest RMSE (5.051).

```python
data = df.drop(columns = ['medv','dis','nox','rad'])
model = ridge_reg_ten

coef_df = pd.DataFrame({
    'feature':['intercept'] + data.columns.tolist(),
    'coefficient':[model.intercept_] + list(model.coef_)
})

coef_df.sort_values('coefficient',ascending=False)
```

>rm and chas has the highest contributions with 5.04 and 2.06. All features seems contribute so little. Most negative value contribution is from ptratio. Interpret :<Br>
>1. 1 value increase of rm is increasing the model predictions by 5.04 with all features remain the same value.<br>
2. if chas is true or 1, it's increasing the model predictions by 2.06 with all features remain the same value.

```python
x_test = x_test.to_numpy()
y_test = y_test.to_numpy()
y_test = y_test.reshape(len(y_test),)

y_predict_test = ridge_reg_ten.predict(x_test)
rmse = np.sqrt(mean_squared_error(y_test ,y_predict_test))
mape1 = mape(y_test ,y_predict_test)
mae1 = mae(y_test ,y_predict_test)
print(f'RMSE of best Ridge regression model on test data set is {rmse}')
print(f'MAPE of best Ridge regression model on test data set is {mape1}')
print(f'MAE of best Ridge regression model on test data set is {mae1}')
```

>In my opinion, the model performs really well on the test data. The RMSE, MAPE and MAE increase not that much by validation data. Let's use it on train data set.

```python
y_predict_train = ridge_reg_ten.predict(x_train)

print('RMSE for training data is {}'.format(np.sqrt(mean_squared_error(y_train ,y_predict_train))))
print('MAE for training data is {}'.format(mae(y_train ,y_predict_train)))
print('MAPE for training data is {}'.format(mape(y_train ,y_predict_train)))
```

> As you can see, the RMSE and MAPE is slightly higher on test data, but the MAE is lower on the test data!. For me, this model is pretty good already, no overfitting and no underfitting.

## Lasso Regression

```python
from sklearn.linear_model import Lasso

# define the model
# 4 models with 4 different alphas (lambda)
lasso_reg_pointzeroone = Lasso(alpha=0.01, random_state=42)
lasso_reg_pointone = Lasso(alpha=0.1, random_state=42)
lasso_reg_one = Lasso(alpha=1, random_state=42)
lasso_reg_ten = Lasso(alpha=10, random_state=42)

# fit the model (training)
lasso_reg_pointzeroone.fit(x_train, y_train)
lasso_reg_pointone.fit(x_train, y_train)
lasso_reg_one.fit(x_train, y_train)
lasso_reg_ten.fit(x_train, y_train)
```

```python
alphas = [0.01, 0.1, 1., 10]
models = [lasso_reg_pointzeroone,
          lasso_reg_pointone,
          lasso_reg_one,
          lasso_reg_ten]

for model, alpha in zip(models, alphas):
    y_predict_validation = model.predict(x_validation)
    rmse = np.sqrt(mean_squared_error(y_validation ,y_predict_validation))
    mape1 = mape(y_validation ,y_predict_validation)
    mae1 = mae(y_validation ,y_predict_validation)
    print(f'RMSE of Lasso regression model with alpha = {alpha} is {rmse}')
    print(f'MAPE of Lasso regression model with alpha = {alpha} is {mape1}')
    print(f'MAE of Lasso regression model with alpha = {alpha} is {mae1}')
    print('--------------------------------------------------')
```

>Best Lasso regression alpha is 1 with RMSE 4.988.

```python
data = df.drop(columns = ['medv','dis','nox','rad'])
model = lasso_reg_one

coef_df = pd.DataFrame({
    'feature':['intercept'] + data.columns.tolist(),
    'coefficient':[model.intercept_] + list(model.coef_)
})

coef_df.sort_values('coefficient',ascending=False)
```
